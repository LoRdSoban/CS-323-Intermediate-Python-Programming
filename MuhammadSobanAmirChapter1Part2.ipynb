```python
#calculate yards_per_attempt and completion_rate
def calc_yards_per_attempt(passing_stats):
    return passing_stats['passing_yards'] / passing_stats['passing_attempts']

def calc_completion_rate(passing_stats):
    return (passing_stats['completed_passes'] / passing_stats['passing_attempts'])*100

ua_passing_stats_sb = {'passing_yards': 321,
                    'passing_attempts': 22,
                    'completed_passes': 15,
                    'touchdowns': 5,
                    'interceptions':  0}

kansas_passing_stats_sb = {'passing_yards': 210,
                    'passing_attempts': 35,
                    'completed_passes': 18,
                    'touchdowns': 0,
                    'interceptions':  2}

ua_passing_stats_ib = {'passing_yards': 343,
                    'passing_attempts': 30,
                    'completed_passes': 20,
                    'touchdowns': 3,
                    'interceptions':  1}

auburn_passing_stats_ib = {'passing_yards': 77,
                    'passing_attempts': 23,
                    'completed_passes': 11,
                    'touchdowns': 1,
                    'interceptions':  0}

print("This program will provide stats for the  2022 Iron Bowl or the 2022 Sugar Bowl\n\n")

print("Please select which game you would like stats for?")
print("1-IB\n2-SB\n")
game_selected = int(input("Enter here: "))

#validation
while (game_selected !=1 and game_selected != 2):
    print("Invalid Input!!\n\n")
    game_selected = int(input("Enter again: "))

print("\nPlease select which team you would like stats for?")

if(game_selected == 1):
    print("1-UA\n2-Auburn\n")
else:
    print("1-UA\n2-Kansas\n")

choice = int(input("Enter here: "))

#selecting the correct dictionary to output
if(game_selected == 1):
    if(choice == 1):
        passing_stats_selected = ua_passing_stats_ib
    else:
        passing_stats_selected = auburn_passing_stats_ib
else:
    if(choice == 1):
        passing_stats_selected = ua_passing_stats_sb
    else:
        passing_stats_selected = kansas_passing_stats_sb



print(f'The passing stats for {"UA" if choice == 1 else ("Auburn" if game_selected == 1 else "Kansas")} are:\n')
        
print(f"Yards: {passing_stats_selected['passing_yards']}")
print(f"Attempts: {passing_stats_selected['passing_attempts']}")  
print(f"Completions: {passing_stats_selected['completed_passes']}")
print(f"Touchdowns: {passing_stats_selected['touchdowns']}")
print(f"Interceptions: {passing_stats_selected['interceptions']}") 
print(f"Yards per attempt: {calc_yards_per_attempt(passing_stats_selected):.2f}")
print(f"Completion percentage: {calc_completion_rate(passing_stats_selected):.1f}%")
```
