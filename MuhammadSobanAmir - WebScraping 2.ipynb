```python
import bs4
import pandas as pd
import time
import requests
```

```python
url = "https://forecast.weather.gov/MapClick.php?lat=33.52072000000004&lon=-86.81178999999997"

r = requests.get(url)

soup = bs4.BeautifulSoup(r.content)

soup
```

```python
seven_day_forecast = soup.find_all(class_='forecast-tombstone')

tonight = seven_day_forecast[0]
print(tonight.prettify())
```

```python
period = tonight.find(class_='period-name').get_text()

short_desc =  tonight.find(class_='short-desc').get_text()

temp =  tonight.find(class_='temp').get_text()

print(period)
print(short_desc)
print(temp)
```

```python
img = tonight.find('img')
desc = img['title']

print(desc)
```

```python
short_desc_tags = soup.select(".tombstone-container .short-desc")

short_descs = [sd.get_text() for sd in short_desc_tags]

print(short_descs)
```

```python
period_tags = soup.select(".tombstone-container .period-name")

periods = [pt.get_text() for pt in period_tags]

print(periods)
```

```python
temp_tags = soup.select(".tombstone-container .temp")

temps = [t.get_text() for t in temp_tags]

print(temps)
```

```python
img_tags = soup.select(".tombstone-container .forecast-icon")

descs = [i['title'] for i in img_tags]

print(descs)
```

```python
weather = pd.DataFrame({
    "periods":periods,
    "short_desc":short_descs,
    "temp":temps,
    "desc":descs
})

weather
```
